```python
import matplotlib.pyplot as plt

# Exemplo de dados
dados = [93,102,110,110,111,111,112,113,113,113,114,114,114,115,116,116,116,116,117,120,120,120,121,121,121,121,122,122,122,123,123,125,126,128,129,130,132,133,138,140]

# Cria o dotplot "manual"
for valor in set(dados):
    # conta quantas vezes aparece
    freq = dados.count(valor)
    # plota pontos empilhados
    plt.plot([valor] * freq, range(1, freq + 1), 'o', color='blue')

plt.xlabel("Valores")
plt.ylabel("Frequência")
plt.title("Dotplot")
plt.show()
```



for valor in set(dados):

- dados tem vários números, alguns repetidos.

- set(dados) pega só os valores únicos (sem repetição).

Assim, o laço percorre cada valor distinto (exemplo: 2, 3, 4, 5, 6, 7).

* *freq = dados.count(valor)*

- Conta quantas vezes aquele valor aparece na lista dados.

- Exemplo: se valor = 6, então freq = 4 (porque o 6 aparece 4 vezes).

[valor] * freq

- Cria uma lista repetindo o valor várias vezes.

- Exemplo: se valor = 6 e freq = 4, vira [6, 6, 6, 6].

- Isso é usado para o eixo X (posição horizontal).

range(1, freq + 1)

- Cria os valores 1, 2, ..., freq.

- Exemplo: se freq = 4, vira 1, 2, 3, 4.

- Isso serve para o eixo Y, empilhando os pontos.

plt.plot([valor] * freq, range(1, freq + 1), 'o', color='blue')

- Faz o gráfico de pontos ('o') no matplotlib.

- Para valor = 6, plota os pontos: (6,1), (6,2), (6,3), (6,4).

- Assim os pontos ficam empilhados verticalmente no valor 6 do eixo X.

<h1>Ramos e Folhas</h1>

```python
import matplotlib.pyplot as plt

def stem_and_leaf_plot(data):
    data = sorted(data)
    stem_dict = {}
    for num in data:
        stem, leaf = divmod(num, 10)
        stem_dict.setdefault(stem, []).append(leaf)

    # Mostrar gráfico
    stems = []
    leaves = []
    for stem, l in stem_dict.items():
        stems.extend([stem] * len(l))
        leaves.extend(l)

    plt.scatter(stems, leaves, marker="o")
    plt.xlabel("Ramo (dezenas)")
    plt.ylabel("Folha (unidades)")
    plt.title("Gráfico de Ramos e Folhas")
    plt.grid(True)
    plt.show()

dados = [93,102,110,110,111,111,112,113,113,113,114,114,114,115,116,116,116,116,117,120,120,120,121,121,121,121,122,122,122,123,123,125,126,128,129,130,132,133,138,140]
stem_and_leaf_plot(dados)
```
